**Going Numeric with Data and removing the units from entries, Collecting units to rename columns**

1. Collecting Units for renaming columns

```python
units=df.astype(str).replace('[^a-zA-Z]','',regex=True)
```

```python
units
```

```python
units=units.mode()
```

```python
units  #collected the units
```

2. Merging units with Column name

```python
for i in units:
  print(units[i].at[0])
```

```python
units=units.replace('',np.nan).dropna(axis=1)
```

```python
for i in units:
  print(units[i].at[0])
```

```python
mapper={i: i+'_'+units[i].at[0] for i in units}
```

```python
mapper
```

```python
df.rename(columns=mapper,inplace=True)
```

3. Removing Units from Values

```python
df.replace('[a-zA-Z]','',regex=True,inplace=True)
```

```python
df=df.astype(float)
```

```python
df.head()
```
